```python
# ãƒãƒ¼ãƒ ã®å‹æ•—ã‚’äºˆæ¸¬ã™ã‚‹ã ã‘ã§ãªãã€é¸æ‰‹ãŒå–œã‚“ã§ã„ã‚‹ã®ã‹ã©ã†ã‹ã‚‚äºˆæ¸¬ã™ã‚‹ã€‚
# ã¾ãŸã€ã‘ãŒã‚’ã—ãŸãƒãƒ¼ãƒ ãƒ¡ãƒ³ãƒãƒ¼ã®å‰²åˆã‚‚äºˆæ¸¬ã™ã‚‹ã€‚ã“ã®äºˆæ¸¬ã§ã¯ã€ç¾åœ¨ã®å‹æ•—è¨˜éŒ²ã®ã¿ã‚’ä½¿ç”¨ã™ã‚‹ã€‚

# äºˆæ¸¬ã‚’ã™ã‚‹é–¢æ•° neural_network ã®å®šç¾©
def neural_network(input, weights):
    pred = ele_mul(input, weights)
    return pred

# åŠ é‡å’Œï¼ˆå†…ç©ï¼‰ã‚’è¨ˆç®—ã™ã‚‹é–¢æ•° ele_mul ã®å®šç¾©
def ele_mul(input, weights):
    # åŠ é‡å’Œãƒªã‚¹ãƒˆã®åˆæœŸåŒ–
    out = []
    # é‡ã¿ãƒªã‚¹ãƒˆã®é•·ã•åˆ†ç¹°ã‚Šè¿”ã™
    for i in range(len(weights)):
        # åŠ é‡å’Œã‚’æ±‚ã‚ã¦ã€åŠ é‡å’Œãƒªã‚¹ãƒˆã«è¿½åŠ ã™ã‚‹
        out.append(input * weights[i])
    # åŠ é‡å’Œã‚’è¿”ã™
    return out

# ã‚¹ã‚«ãƒ©ãƒ¼ç©ã‚’æ±‚ã‚ã‚‹é–¢æ•° scalar_ele_mul ã®å®šç¾©
def scalar_ele_mul(number, vector):
    # ã‚¹ã‚«ãƒ©ãƒ¼ç©ã‚’è¦ç´ ã¨ã™ã‚‹ãƒªã‚¹ãƒˆã®åˆæœŸåŒ–
    output = []
    # å¼•æ•° vector ã®é•·ã•åˆ†ç¹°ã‚Šè¿”ã™
    for i in range(len(vector)):
        # ã‚¹ã‚«ãƒ©ãƒ¼ç©ã‚’æ±‚ã‚ã¦ãƒªã‚¹ãƒˆã«è¿½åŠ ã™ã‚‹
        output.append(number * vector[i])
    # ã‚¹ã‚«ãƒ©ãƒ¼ç©ã‚’è¦ç´ ã¨ã™ã‚‹ãƒªã‚¹ãƒˆã‚’è¿”ã™
    return output

# é‡ã¿ã®åˆæœŸåŒ–ï¼š[å‹æ•—ã‹ã‚‰ã‘ãŒï¼Ÿã¸ã®é‡ã¿, å‹æ•—ã‹ã‚‰å‹ã£ãŸï¼Ÿã¸ã®é‡ã¿, å‹æ•—ã‹ã‚‰æ‚²ã—ã„ï¼Ÿã¸ã®é‡ã¿]
weights = [0.3, 0.2, 0.9]

# ï¼‘ã‚·ãƒ¼ã‚ºãƒ³ï¼”è©¦åˆã®å‹ç‡ã®åˆæœŸåŒ–
wlrec = [0.65, 1.0, 1.0, 0.9]
# ï¼‘ã‚·ãƒ¼ã‚ºãƒ³ï¼”è©¦åˆã®ã‘ãŒã®åˆæœŸåŒ–
hurt  = [0.1, 0.0, 0.0, 0.1]
# ï¼‘ã‚·ãƒ¼ã‚ºãƒ³ï¼”è©¦åˆã®å‹ã¡è² ã‘ã®åˆæœŸåŒ–
win   = [  1,   1,   0,   1]
# ï¼‘ã‚·ãƒ¼ã‚ºãƒ³ï¼”è©¦åˆã®æ‚²ã—ã¿åº¦ã®åˆæœŸåŒ–
sad   = [0.1, 0.0, 0.1, 0.2]

# å…¥åŠ›å€¤ã®åˆæœŸåŒ–ï¼ˆã‚·ãƒ¼ã‚ºãƒ³ç¬¬ï¼‘è©¦åˆã®å‹ç‡ï¼‰
input = wlrec[0]
# ã‚·ãƒ¼ã‚ºãƒ³ç¬¬ï¼‘è©¦åˆã®äº‹å®Ÿ
# [ç¬¬ï¼‘è©¦åˆã®ã‘ãŒ, ç¬¬ï¼‘è©¦åˆã®å‹ã¡è² ã‘, ç¬¬ï¼‘è©¦åˆã®æ‚²ã—ã¿åº¦]
truth = [hurt[0], win[0], sad[0]]

# äºˆæ¸¬ã™ã‚‹
pred = neural_network(input, weights)

# ãƒ‡ãƒ«ã‚¿ï¼ˆç´”èª¤å·®ï¼‰ãƒªã‚¹ãƒˆã®åˆæœŸåŒ–
# [ã‘ãŒ, å‹ã¡è² ã‘, æ‚²ã—ã¿åº¦]
delta = []
# å‡ºåŠ›å±¤ã®ãƒ¦ãƒ‹ãƒƒãƒˆæ•°åˆ†ç¹°ã‚Šè¿”ã™
for i in range(len(truth)):
    # å„ãƒ¦ãƒ‹ãƒƒãƒˆã”ã¨ã®ç´”èª¤å·®ã‚’æ±‚ã‚ã¦ã€ç´”èª¤å·®ãƒªã‚¹ãƒˆã«è¿½åŠ ã™ã‚‹
    delta.append(pred[i] - truth[i])

# é‡ã¿å¾®èª¿æ•´é‡ã‚’æ±‚ã‚ã‚‹
weight_deltas = scalar_ele_mul(input, delta)
# é‡ã¿å¾®èª¿æ•´ç”¨ã‚¢ãƒ«ãƒ•ã‚¡ã®åˆæœŸåŒ–
alpha = 0.1
# å­¦ç¿’ã™ã‚‹
for i in range(len(weights)):
    # é‡ã¿ã‚’æ›´æ–°ã™ã‚‹
    weights[i] -= (weight_deltas[i] * alpha)

# é‡ã¿ã‚’è¡¨ç¤º
print("Weights:" + str(weights))

# äºˆæ¸¬å€¤ã®è¡¨ç¤º
print('[ã‘ãŒåº¦ = {}, å‹ã¡è² ã‘ = {}, æ‚²ã—ã¿åº¦ = {}]'.format(pred[0], pred[1], pred[2]))
```

### ğŸ‘‡æœªå®Œæˆãƒªã‚¹ãƒˆ

```python
# ãƒãƒ¼ãƒ ã®å‹æ•—ã‚’äºˆæ¸¬ã™ã‚‹ã ã‘ã§ãªãã€é¸æ‰‹ãŒå–œã‚“ã§ã„ã‚‹ã®ã‹ã©ã†ã‹ã‚‚äºˆæ¸¬ã™ã‚‹ã€‚
# ã¾ãŸã€ã‘ãŒã‚’ã—ãŸãƒãƒ¼ãƒ ãƒ¡ãƒ³ãƒãƒ¼ã®å‰²åˆã‚‚äºˆæ¸¬ã™ã‚‹ã€‚ã“ã®äºˆæ¸¬ã§ã¯ã€ç¾åœ¨ã®å‹æ•—è¨˜éŒ²ã®ã¿ã‚’ä½¿ç”¨ã™ã‚‹ã€‚

# äºˆæ¸¬ã‚’ã™ã‚‹é–¢æ•° neural_network ã®å®šç¾©
def neural_network(input, weights):
    pred = ele_mul(input, weights)
    return pred

# åŠ é‡å’Œï¼ˆå†…ç©ï¼‰ã‚’è¨ˆç®—ã™ã‚‹é–¢æ•° ele_mul ã®å®šç¾©
def ele_mul(input, weights):
    # åŠ é‡å’Œãƒªã‚¹ãƒˆã®åˆæœŸåŒ–
    out = []
    # é‡ã¿ãƒªã‚¹ãƒˆã®é•·ã•åˆ†ç¹°ã‚Šè¿”ã™
    for i in range(len(weights)):
        # åŠ é‡å’Œã‚’æ±‚ã‚ã¦ã€åŠ é‡å’Œãƒªã‚¹ãƒˆã«è¿½åŠ ã™ã‚‹
        out.append(input * weights[i])
    # åŠ é‡å’Œã‚’è¿”ã™
    return out

# ã‚¹ã‚«ãƒ©ãƒ¼ç©ã‚’æ±‚ã‚ã‚‹é–¢æ•° scalar_ele_mul ã®å®šç¾©
def scalar_ele_mul(number, vector):
    # ã‚¹ã‚«ãƒ©ãƒ¼ç©ã‚’è¦ç´ ã¨ã™ã‚‹ãƒªã‚¹ãƒˆã®åˆæœŸåŒ–
    output = []
    # å¼•æ•° vector ã®é•·ã•åˆ†ç¹°ã‚Šè¿”ã™
    for i in range(len(vector)):
        # ã‚¹ã‚«ãƒ©ãƒ¼ç©ã‚’æ±‚ã‚ã‚‹
        output.append(number * vector[i])
    # ã‚¹ã‚«ãƒ©ãƒ¼ç©ã‚’è¦ç´ ã¨ã™ã‚‹ãƒªã‚¹ãƒˆã‚’è¿”ã™
    return output

# å­¦ç¿’é–¢æ•° grad_descent_learn(input, truth, pred, weights, alpha) ã®å®šç¾©
'''
é–¢æ•°åï¼šgrad_descent_learn
å¼•æ•°ï¼š
    inputï¼šå…¥åŠ›å€¤
    truthï¼šç›®çš„å€¤ãƒªã‚¹ãƒˆ
    predï¼šäºˆæ¸¬å€¤ãƒªã‚¹ãƒˆ
    weightsï¼šé‡ã¿ãƒªã‚¹ãƒˆ
    alphaï¼šé‡ã¿å†å¾®èª¿æ•´å€¤
å‡¦ç†ï¼šå‹¾é…é™ä¸‹æ³•ã«åŸºã¥ãé‡ã¿ã‚’ä¿®æ­£ã™ã‚‹
æˆ»ã‚Šå€¤ï¼šä¿®æ­£ã•ã‚ŒãŸé‡ã¿ãƒªã‚¹ãƒˆ
'''

def grad_descent_learn(input, truth, pred, weights, alpha):
    # ãƒ‡ãƒ«ã‚¿ã®åˆæœŸåŒ–
    delta = []
    # èª¤å·®ã¨ãƒ‡ãƒ«ã‚¿ã‚’æ±‚ã‚ã‚‹
    for i in range(len(truth)):
        delta.append(pred[i] - truth[i])
    weights_deltas = scalar_ele_mul(input, delta)
    for i in range(len(weights)):
        weights[i] -= alpha * weights_deltas[i]
    return weights

# ï¼‘ã‚·ãƒ¼ã‚ºãƒ³ï¼”è©¦åˆã®å‹ç‡ã®åˆæœŸåŒ–
wlrec = [0.65, 1.0, 1.0, 0.9]
# ï¼‘ã‚·ãƒ¼ã‚ºãƒ³ï¼”è©¦åˆã®ã‘ãŒã®åˆæœŸåŒ–
hurt  = [0.1, 0.0, 0.0, 0.1]
# ï¼‘ã‚·ãƒ¼ã‚ºãƒ³ï¼”è©¦åˆã®å‹ã¡è² ã‘ã®åˆæœŸåŒ–
win   = [  1,   1,   0,   1]
# ï¼‘ã‚·ãƒ¼ã‚ºãƒ³ï¼”è©¦åˆã®æ‚²ã—ã¿åº¦ã®åˆæœŸåŒ–
sad   = [0.1, 0.0, 0.1, 0.2]
# é‡ã¿ã®åˆæœŸåŒ–ï¼š[å‹æ•—ã‹ã‚‰ã‘ãŒï¼Ÿã¸ã®é‡ã¿, å‹æ•—ã‹ã‚‰å‹ã£ãŸï¼Ÿã¸ã®é‡ã¿, å‹æ•—ã‹ã‚‰æ‚²ã—ã„ï¼Ÿã¸ã®é‡ã¿]
weights = [0.3, 0.2, 0.9] 
# å…¥åŠ›å€¤ã®åˆæœŸåŒ–ï¼ˆã‚·ãƒ¼ã‚ºãƒ³ç¬¬ï¼‘è©¦åˆã®å‹ç‡ï¼‰
input = wlrec[0]
# ã‚·ãƒ¼ã‚ºãƒ³ç¬¬ï¼‘è©¦åˆã®äº‹å®Ÿ
# [ç¬¬ï¼‘è©¦åˆã®ã‘ãŒ, ç¬¬ï¼‘è©¦åˆã®å‹ã¡è² ã‘, ç¬¬ï¼‘è©¦åˆã®æ‚²ã—ã¿åº¦]
truth = [hurt[0], win[0], sad[0]]

# äºˆæ¸¬ã™ã‚‹
pred = neural_network(input, weights)
# é‡ã¿å¾®èª¿æ•´ç”¨ã‚¢ãƒ«ãƒ•ã‚¡ã®åˆæœŸåŒ–
alpha = 0.1
# å­¦ç¿’ã™ã‚‹
weights = grad_descent_learn(input, truth, pred, weights, alpha)

# é‡ã¿ã‚’è¡¨ç¤º
print("Weights:" + str(weights))

# äºˆæ¸¬å€¤ã®è¡¨ç¤º
print('[ã‘ãŒåº¦ = {}, å‹ã¡è² ã‘ = {}, æ‚²ã—ã¿åº¦ = {}]'.format(pred[0], pred[1], pred[2]))
```
