```python
# NumPyã®ã‚¤ãƒ³ãƒãƒ¼ãƒˆã€åˆ¥å np
import numpy as np

# åŠ é‡å’Œï¼ˆå†…ç©ï¼‰ã‚’æ±‚ã‚ã‚‹é–¢æ•° w_sum ã®å®šç¾©
def w_sum(a, b):
    # ï¼’ã¤ã®å¼•æ•°ã®ãƒªã‚¹ãƒˆã®é•·ã•ãŒç­‰ã—ã„ã¨ãä»¥ä¸‹ã‚’å®Ÿè¡Œã™ã‚‹
    assert(len(a) == len(b))
    # åŠ é‡å’Œã®åˆæœŸåŒ–
    output = 0
    # åŠ é‡å’Œã‚’æ±‚ã‚ã‚‹
    # å¼•æ•°ã®ãƒªã‚¹ãƒˆã®é•·ã•åˆ†ç¹°ã‚Šè¿”ã™
    for i in range(len(a)):
        # ãƒªã‚¹ãƒˆã®è¦ç´ åŒå£«ã‚’æ›ã‘ã¦åˆè¨ˆã‚’æ±‚ã‚ã‚‹
        output += (a[i] * b[i])
    # åŠ é‡å’Œã‚’è¿”ã™
    return output

# ãƒ™ã‚¯ãƒˆãƒ«ã¨è¡Œåˆ—ã®ã‹ã‘ç®—ã‚’æ±‚ã‚ã‚‹é–¢æ•° vect_mat_mul ã®å®šç¾©
def vect_mat_mul(vect, matrix):
    # ãƒ™ã‚¯ãƒˆãƒ«ã¨è¡Œåˆ—ã®è¡Œãƒªã‚¹ãƒˆã®é•·ã•ãŒç­‰ã—ã„ã¨ãä»¥ä¸‹ã‚’å®Ÿè¡Œã™ã‚‹
    assert(len(vect) == len(matrix))
    # åŠ é‡å’Œãƒªã‚¹ãƒˆã®åˆæœŸåŒ–
    out = []
    # ãƒ™ã‚¯ãƒˆãƒ«ã®é•·ã•ï¼ˆï¼è¡Œåˆ—ã®è¡Œãƒªã‚¹ãƒˆã®é•·ã•ï¼‰åˆ†ç¹°ã‚Šè¿”ã™
    for i in range(len(matrix)):
        # ãƒ™ã‚¯ãƒˆãƒ«ã®è¡Œåˆ—ã®å„è¡Œã¨ã®åŠ é‡å’Œã‚’æ±‚ã‚ã¦ out ã«è¿½åŠ ã™ã‚‹
        out.append(w_sum(vect, matrix[i]))
    # åŠ é‡å’Œã®ãƒªã‚¹ãƒˆã‚’è¿”ã™
    return out

# ï¼’ã¤ã®ãƒªã‚¹ãƒˆã®è¦ç´ é–“ã‚’ç·ãªã‚ã—ã¦è¦ç´ ã”ã¨ã®ã‹ã‘ç®—ã‚’ã—ã¦
# ãã®çµæœã‚’è¡Œåˆ—ã«æ ¼ç´ã™ã‚‹é–¢æ•° outer_prod ã®å®šç¾©
def outer_prod(a, b):
    # a è¡Œ b åˆ—ã®ã‚¼ãƒ­è¡Œåˆ—ã‚’ç”Ÿæˆã™ã‚‹
    out = np.zeros((len(a), len(b)))
    # a è¡Œ b åˆ—ã®è¡Œåˆ—ã«ãƒªã‚¹ãƒˆ a ã¨ãƒªã‚¹ãƒˆ b ã®
    # è¦ç´ ã‚’ç·ãªã‚ã—ãŸã‹ã‘ç®—ã‚’è¡Œã„ã€è¡Œåˆ—ã«æ ¼ç´ã™ã‚‹
    for i in range(len(a)):
        for j in range(len(b)):
            out[i][j] = a[i] * b[j]
    # è¦ç´ é–“ã®ã‹ã‘ç®—ã®çµæœã‚’ä¿å­˜ã—ãŸè¡Œåˆ—ã‚’è¿”ã™
    return out

# äºˆæ¸¬å€¤ã‚’æ±‚ã‚ã‚‹é–¢æ•° nuural_network ã®å®šç¾©
def neural_network(input, weights):
    # ãƒ™ã‚¯ãƒˆãƒ«ã¨è¡Œåˆ—ã®åŠ é‡å’Œã‚’æ±‚ã‚ã‚‹
    pred = vect_mat_mul(input, weights)
    # äºˆæ¸¬å€¤ã®ãƒªã‚¹ãƒˆã‚’è¿”ã™
    return pred

# å­¦ç¿’é–¢æ•° grad_descent_learn(input, truth, pred, weights, alpha) ã®å®šç¾©
'''
é–¢æ•°åï¼šgrad_descent_learn
å¼•æ•°ï¼š
    inputï¼šå…¥åŠ›å€¤ãƒªã‚¹ãƒˆ
    truthï¼šç›®çš„å€¤ãƒªã‚¹ãƒˆ
    predï¼šäºˆæ¸¬å€¤ãƒªã‚¹ãƒˆ
    weightsï¼šé‡ã¿ãƒªã‚¹ãƒˆ
    alphaï¼šé‡ã¿å†å¾®èª¿æ•´å€¤
å‡¦ç†ï¼šå‹¾é…é™ä¸‹æ³•ã«åŸºã¥ãé‡ã¿ã‚’ä¿®æ­£ã™ã‚‹
æˆ»ã‚Šå€¤ï¼šä¿®æ­£ã•ã‚ŒãŸé‡ã¿ãƒªã‚¹ãƒˆ
'''
def grad_descent_learn(input, truth, pred, weights, alpha):
    # ãƒ‡ãƒ«ã‚¿ã®åˆæœŸåŒ–
    delta = []
    # äºˆæ¸¬å€¤ã®æ•°åˆ†ç¹°ã‚Šè¿”ã™
    for i in range(len(pred)):
        # ãƒ‡ãƒ«ã‚¿ã‚’æ±‚ã‚ã¦ãƒ‡ãƒ«ã‚¿ã®ãƒªã‚¹ãƒˆã«è¿½åŠ ã™ã‚‹
        delta.append(pred[i] - truth[i])
     # é‡ã¿ã®å¾®èª¿æ•´é‡ã‚’æ±‚ã‚è¡Œåˆ—ã«æ ¼ç´ã™ã‚‹
    weight_deltas = other_prod(delta, input)
    # é‡ã¿è¡Œåˆ—ã‚’æ›´æ–°ã™ã‚‹ï¼ˆï¼’é‡ãƒ«ãƒ¼ãƒ—ï¼‰
    for i in range(len(weights)):
        for j in range(len(weights[i])):
                weights[i][j] -= alpha * weight_deltas[i][j]

    # é‡ã¿è¡Œåˆ—ã‚’è¿”ã™
    return weights


# é‡ã¿ã®åˆæœŸåŒ–
            #toes %win #fans
weights = [ [0.1, 0.1, -0.3], # ã‘ãŒï¼Ÿ
            [0.1, 0.2, 0.0],  # å‹ã£ãŸï¼Ÿ
            [0.0, 1.3, 0.1] ] # æ‚²ã—ã„ï¼Ÿ

# ã‚·ãƒ¼ã‚ºãƒ³ï¼”è©¦åˆã®è¶³æŒ‡ã®æ•°ã®å¹³å‡
toes  = [8.5, 9.5, 9.9, 9.0]
# ã‚·ãƒ¼ã‚ºãƒ³ï¼”è©¦åˆã®å‹ç‡
wlrec = [0.65,0.8, 0.8, 0.9]
# ã‚·ãƒ¼ã‚ºãƒ³ï¼”è©¦åˆã®ãƒ•ã‚¡ãƒ³ã®æ•°
nfans = [1.2, 1.3, 0.5, 1.0]

# ã‚·ãƒ¼ã‚ºãƒ³ï¼”è©¦åˆã®ã‘ãŒåº¦
hurt  = [0.1, 0.0, 0.0, 0.1]
# ã‚·ãƒ¼ã‚ºãƒ³ï¼”è©¦åˆã®å‹ã¡è² ã‘
win   = [  1,   1,   0,   1]
# ã‚·ãƒ¼ã‚ºãƒ³ï¼”è©¦åˆã®æ‚²ã—ã¿åº¦
sad   = [0.1, 0.0, 0.1, 0.2]
# ã‚¢ãƒ«ãƒ•ã‚¡ã®åˆæœŸåŒ–
alpha = 0.01
# å…¥åŠ›ãƒ‡ãƒ¼ã‚¿ã®è¨­å®šï¼ˆã‚·ãƒ¼ã‚ºãƒ³ç¬¬ï¼‘è©¦åˆã® [è¶³æŒ‡ã®æ•°ã®å¹³å‡å€¤, å‹ç‡, ãƒ•ã‚¡ãƒ³ã®æ•°]
input = [toes[0], wlrec[0], nfans[0]]
# ã‚·ãƒ¼ã‚ºãƒ³ç¬¬ï¼‘è©¦åˆã®çµæœã®è¨­å®šï¼ˆ[ã‘ãŒåº¦, å‹ã¡è² ã‘, æ‚²ã—ã¿åº¦]
truth  = [hurt[0], win[0], sad[0]]

# å‹¾é…é™ä¸‹æ³•ã«ã‚ˆã‚‹å­¦ç¿’
# äºˆæ¸¬å€¤ã‚’æ±‚ã‚ã‚‹
pred = neural_network(input, weights)
# å­¦ç¿’ã™ã‚‹
weights = grad_descent_learn(input, truth, pred, weights, alpha)

# äºˆæ¸¬å€¤ã‚’è¡¨ç¤º
print("äºˆæ¸¬å€¤ = " + str(pred))
print('æ›´æ–°ã•ã‚ŒãŸé‡ã¿')
# æ›´æ–°ã•ã‚ŒãŸé‡ã¿è¡Œåˆ—ã‚’è¡¨ç¤ºã™ã‚‹ï¼ˆãƒ«ãƒ¼ãƒ—ï¼‰
for i in range(len(weights)):
    print(weights[i][0])
```
